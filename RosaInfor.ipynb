```python
def estpremier(n):
    i=2
    rep=True
    while rep and i<=n/2:
        
        if n%i==0: 
            rep=False
            
        i+=1
        
    return rep       
    
```

```python
for i in range(1,200):
    if estpremier(i):
        print(i,end=";")
```

2019 n'est pas premier! La somme des chiffres de 2019 est divisible par 3.
        
    

```python
def sommedeschiffresb(n):
    #Conversion en binaire
    #La conversion donne une chaine de caractère qui commence par 0b
    nb=bin(n)
    S=0
    for i in range(2,len(nb)):
        S=S+int(nb[i])
    #On retourne la somme des chiffres du nb binaire
    return S
    
```

```python
sommedeschiffresb(2019)
```

```python
def election(n):
    #n=int(input("année en cours"))
    if n%2==1:
        if not(estpremier(n)):
            print(n," election")
        
        elif sommedeschiffresb(n)==9:
            print(n," election")
        
```

```python
for i in range(2019,2059):
    election(i)
```
